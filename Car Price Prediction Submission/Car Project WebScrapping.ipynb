```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
from webdriver_manager.chrome import ChromeDriverManager
import re
```

```python
driver = webdriver.Chrome('chromedriverv99.exe')
time.sleep(2)

url = "https://www.cars24.com/"
driver.get(url)
driver.maximize_window()
```

```python
selectmanually = driver.find_element_by_xpath("/html/body/div[1]/div[10]/div/div/div/div[2]/div[2]/button")
selectmanually
selectmanually.click()

time.sleep(2)
```

```python
selectcity = driver.find_element_by_xpath("/html/body/div[1]/div[10]/div/div/div/div[2]/ul/li[1]/span/img")
selectcity
selectcity.click()

time.sleep(2)
```

```python
viewcars = driver.find_element_by_xpath("/html/body/div[1]/div[1]/header/div/div[2]/div/span")
viewcars
viewcars.click()

for _ in range(170):
    driver.execute_script("window.scrollBy(0,170)")
time.sleep(5)
```

```python
urls = []

urldriver = driver.find_elements_by_xpath("//div[@class='col-4']/a")

for link in urldriver:
    urls.append(link.get_attribute('href'))
urls
```

```python
carbrand = []
yearofmake = []
price = []
kilometersdriven = []
transmission = []
fueltype = []
owner = []


for link in urls:                 
    driver.get(link)
    time.sleep(2)
    
    try:                    
        cbrand =driver.find_element_by_xpath("//h2[@class='_2geSF']")
        carbrand.append(cbrand.text.split(" ")[1])
    except NoSuchElementException:
        carbrand.append('No details available')
        
    try:                    
        year =driver.find_element_by_xpath("//h2[@class='_2geSF']")
        yearofmake.append(year.text.split(" ")[0])
    except NoSuchElementException:
        yearofmake.append('No details available')

    try:                    
        prce =driver.find_element_by_xpath("//div[@class='text-right g-hPT']/strong")
        price.append(prce.text.strip("Fixed Price\n"))
    except NoSuchElementException:
        price.append('No details available')        

    try:                    
        kmd =driver.find_element_by_xpath("//ul[@class='_2KO13']/li[3]/strong")
        kilometersdriven.append(kmd.text)
    except NoSuchElementException:
        kilometersdriven.append('No details available')

    try:                    
        trans =driver.find_element_by_xpath("//ul[@class='_2KO13']/li[6]/strong")
        transmission.append(trans.text)
    except NoSuchElementException:
        transmission.append('No details available')        
        
    try:                    
        ftype =driver.find_element_by_xpath("//ul[@class='_2KO13']/li[4]/strong")
        fueltype.append(ftype.text)
    except NoSuchElementException:
        fueltype.append('No details available')

    try:                    
        own =driver.find_element_by_xpath("//ul[@class='_2KO13']/li[2]/strong")
        owner.append(own.text)
    except NoSuchElementException:
        owner.append('No details available')        
        
        
        
usercarlist = pd.DataFrame()
usercarlist['carbrandname'] = carbrand
usercarlist['yearofthemake'] = yearofmake
usercarlist['price'] = price
usercarlist['kilometersdriven'] = kilometersdriven
usercarlist['transmission'] = transmission
usercarlist['fueltype'] = fueltype
usercarlist['owner'] = owner
usercarlist
```

```python
usercarlist.to_csv("usercarlist.csv")
```
