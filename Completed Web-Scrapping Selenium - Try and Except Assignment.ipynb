```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
amazon = webdriver.Chrome('chromedriver.exe')
time.sleep(2)
url = "https://www.amazon.in/"
amazon.get(url)
time.sleep(2)

userinput = input('Enter the Search Keyword: ')
```

```python
search = amazon.find_element_by_id("twotabsearchtextbox")
search
time.sleep(2)
search.send_keys(userinput)
time.sleep(2)
searchbtn = amazon.find_element_by_id("nav-search-submit-text")
searchbtn
searchbtn.click()
```

```python
product_urls = []

for page in range(0,3):#for loop for scrapping 3 page
    url=amazon.find_elements_by_xpath('//a[@class="a-link-normal s-no-outline"]')       
    for i in url:
        product_urls.append(i.get_attribute("href"))                       
    nextpage = amazon.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nextpage.click()
    time.sleep(2)
product_urls
```

```python
len(product_urls)
```

```python
brandname = []
productname = []
price = []
rating = []

for link in product_urls[0:5]:
    amazon.get(link)
    time.sleep(1)
    
    try:
        brand = amazon.find_element_by_xpath('//*[@id="bylineInfo"]')
        brandname.append(brand.text)
    except NoSuchElementException:
        brandname.append('-')
        
    time.sleep(2)
    
    try:
        title = amazon.find_element_by_id("productTitle")
        productname.append(title.text)
    except NoSuchElementException:
        productname.append('-')

    time.sleep(2)
    
    try:
        prc = amazon.find_element_by_xpath('//td[@class="a-span12"]')
        price.append(prc.text)
    except NoSuchElementException:
        price.append('-')

    time.sleep(2)
    
    try:
        rat = amazon.find_element_by_xpath('//span[@class="a-size-base a-nowrap"]')
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append('-')

finaldata = pd.DataFrame()
finaldata ['Brand Name'] = brandname
finaldata ['Product Name'] = productname
finaldata ['Price'] = price
finaldata ['Rating'] = rating
finaldata ['Product URL'] = product_urls[0:5]
finaldata.head(11)
```

```python
urls = fkart.find_elements_by_xpath('//a[@class="_1fQZEK"]')
urls

fkartproducturls = []

for url in urls:
    fkartproducturls.append(url.get_attribute("href"))
fkartproducturls
```

```python
brandname = []
phonename = []
colour = []
storage = []
ram = []
primarycamera = []
secondarycamera = []
displaysize = []
batterycapacity = []
price = []


for links in fkartproducturls[0:10]:
    fkart.get(links)

    try:
        name = fkart.find_element_by_xpath('//h1[@class="yhB1nd"]')
        brandname.append(name.text.split(" ")[0])
    except NoSuchElementException:
        brandname.append('-')

    try:
        phname = fkart.find_element_by_xpath('//h1[@class="yhB1nd"]')
        phonename.append(phname.text)
    except NoSuchElementException:
        phonename.append('-')

    try:
        clr = fkart.find_element_by_xpath('//table[@class="_14cfVK"]//tr[4]//td[2]')
        colour.append(clr.text)
    except NoSuchElementException:
        colour.append('-')

        
    try: #for expanding readmore button
        read_more_btn = fkart.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    except NoSuchElementException:
        pass
    time.sleep(3)

    try:
        strom = fkart.find_element_by_xpath("//div[@class='_1UhVsV']//div[4]//tr[1]//td[2]")
        storage.append(strom.text)
    except NoSuchElementException:
        storage.append('-')

    try:
        phram = fkart.find_element_by_xpath("//div[@class='_1UhVsV']//div[4]//tr[2]//td[2]")
        ram.append(phram.text)
    except NoSuchElementException:
        ram.append('-')
                
    try:
        pcamera = fkart.find_element_by_xpath("//div[@class='_1UhVsV']//div[5]//tr[2]//td[2]")
        primarycamera.append(pcamera.text)
    except NoSuchElementException:
        primarycamera.append('-')
        
    try:
        scamera = fkart.find_element_by_xpath("//div[@class='_1UhVsV']//div[5]//tr[5]//td[2]")
        secondarycamera.append(scamera.text)
    except NoSuchElementException:
        secondarycamera.append('-')
        
    try:
        dsize = fkart.find_element_by_xpath("//div[@class='_1UhVsV']//div[2]//tr[1]//td[2]")
        displaysize.append(dsize.text)
    except NoSuchElementException:
        displaysize.append('-')
        
    try:
        bcapacity = fkart.find_element_by_xpath("//div[@class='_1UhVsV']//div[10]//tr[1]//td[2]")
        batterycapacity.append(bcapacity.text)
    except NoSuchElementException:
        batterycapacity.append('-')

    try:
        rate = fkart.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        price.append(rate.text)
    except NoSuchElementException:
        price.append('-')
        
data = pd.DataFrame()
data['Brand Name'] = brandname
data['Phone Name'] = phonename
data['Price'] = price
data ['Colour'] = colour
data ['ROM/Storage'] = storage
data ['RAM'] = ram
data ['Primary Camera'] = primarycamera
data ['Secondary Camera'] = secondarycamera
data ['Display Size'] = displaysize
data ['Battery Size'] = batterycapacity
data ['Product URL'] = fkartproducturls[0:10]
data.head()
```

```python
hostelurl = []
propdesc = []

while(True):
    urls = hworld.find_elements_by_xpath('//h2[@class="title title-6"]//a')
    for link in urls:
        hostelurl.append(link.get_attribute("href"))
    time.sleep(2)    
        
    try:
        next_button = hworld.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break


```

```python
names = []
distance = []
review = []
rating = []
totalreview = []
facilities = []
combinedprices = []

while(True):
    hostelnames = hworld.find_elements_by_xpath('//h2[@class="title title-6"]')
    for name in hostelnames:
            names.append(name.text)
    
    hosteldistance = hworld.find_elements_by_xpath('//span[@class="description"]')
    for dist in hosteldistance:
            distance.append(dist.text.split("-")[1])
            
    hostelreview = hworld.find_elements_by_xpath('//div[@class="keyword"]')
    for rev in hostelreview:
            review.append(rev.text)     

    hosteltotalreview = hworld.find_elements_by_xpath('//div[@class="reviews"]')
    for treview in hosteltotalreview:
            totalreview.append(treview.text)
            
    hostelhmepagefacilities = hworld.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
    for facility in hostelhmepagefacilities:
            facilities.append(facility.text.replace("\n", ","))
    
    hostelprivateformprices = hworld.find_elements_by_xpath("//div[@class='price-col']")
    for prices in hostelprivateformprices:
            combinedprices.append(prices.text)
    
    
    try:
        next_button = hworld.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
        
combinedprices

```

```python
propdesc = []
overallratings = []
for hotel in hostelurl:
    hworld.get(hotel)


    try:
        propertydescription = hworld.find_element_by_xpath('//div[@class="content collapse-content"]')
        propdesc.append(propertydescription.text.replace("\n", " "))
    except NoSuchElementException:
        propdesc.append("Description not available")

    try:
        overallating = hworld.find_element_by_xpath('//div[@class="keyword"]')
        overallratings.append(overallating.text)
    except NoSuchElementException:
        overallratings.append("overall Rating not Available")

overallratings
```

```python
Privateroomprice = []
for price in range(0,len(combinedprices),2):
    Privateroomprice.append(combinedprices[price].strip("Privates From\n"))
Privateroomprice

dormroomprice = []
for dprice in range(1,len(combinedprices),2):
    dormroomprice.append(combinedprices[dprice].replace("\n", " "))
dormroomprice

Privateroomprice
```
