```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
```

```python
iccwteam = requests.get(url)
iccwteam
```

```python
iccwteam.content
soup = BeautifulSoup(iccwteam.content, 'html.parser')
print(soup.prettify())
```

```python
iccwteam = soup.find_all('span', class_="u-hide-phablet")
iccwteam
```

```python
WomensteamODIRanking = []

for wodi in iccwteam:
    wodi = wodi.get_text()
    WomensteamODIRanking.append(wodi)
WomensteamODIRanking

WODITeamRanking=list(WomensteamODIRanking)[:10]
WODITeamRanking
```

```python
iccwteam1 = soup.find_all('td', class_="rankings-block__banner--matches")
iccwteam1
```

```python
wteam1matches = []

for team1 in iccwteam1:
    team1 = team1.get_text()
    wteam1matches.append(team1)
wteam1matches
```

```python
iccwteam2 = soup.find_all('tr', class_="table-body")
iccwteam2
```

```python
wteam2to20matches = []

for wteam2 in iccwteam2:
    wteam2 = wteam2.get_text().split('\n')[7]
    wteam2to20matches.append(wteam2)
wteam2to20matches
```

```python
WODITeamMatches = (wteam1matches+wteam2to20matches)[:10]
WODITeamMatches
```

```python
iccwteam3 = soup.find_all('td', class_="rankings-block__banner--points")
iccwteam3
```

```python
wteam3points = []

for wteam3 in iccwteam3:
    wteam3 = wteam3.get_text()
    wteam3points.append(wteam3)
wteam3points
```

```python
iccwteam4 = soup.find_all('tr', class_="table-body")
iccwteam4
```

```python
wteam2to20points = []

for wteam4 in iccwteam4:
    wteam4 = wteam4.get_text().split('\n')[8]
    wteam2to20points.append(wteam4)
wteam2to20points
```

```python
WODITeamPoints = (wteam3points+wteam2to20points)[:10]
WODITeamPoints
```

```python
iccwteam5 = soup.find_all('td', class_="rankings-block__banner--rating u-text-right")
iccwteam5
```

```python
wteam5points = []

for wteam5 in iccwteam5:
    wteam5 = wteam5.get_text().lstrip().split('\n')[0]
    wteam5points.append(wteam5)
wteam5points
```

```python
iccwteam6 = soup.find_all('tr', class_="table-body")
iccwteam6
```

```python
wteam2to20rating = []

for wteam6 in iccwteam6:
    wteam6 = wteam6.get_text().split('\n')[9]
    wteam2to20rating.append(wteam6)
wteam2to20rating
```

```python
WODITeamRating = (wteam5points+wteam2to20rating)[:10]
WODITeamRating
```

```python
data = pd.DataFrame()
data['Teams'] = WODITeamRanking
data['Matches'] = WODITeamMatches
data['Points'] = WODITeamPoints
data['Rating'] = WODITeamRating
data.head(20)
```

```python
data.to_csv('ICC Womens ODI Ranking', index=False)
```
