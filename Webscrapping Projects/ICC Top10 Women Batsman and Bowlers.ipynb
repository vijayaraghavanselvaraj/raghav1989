```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.icc-cricket.com/rankings/womens/player-rankings/odi'
```

```python
top10womenbatsman = requests.get(url)
top10womenbatsman
```

```python
top10womenbatsman.content
soup = BeautifulSoup (top10womenbatsman.content, 'html.parser')
print(soup.prettify())
```

```python
top10womenbatsman1 = soup.find_all('div', class_="rankings-block__banner--name")
top10womenbatsman1
```

```python
topwbatsman1 = []

for tpbats1 in top10womenbatsman1:
    tpbats1 = tpbats1.get_text()
    topwbatsman1.append(tpbats1)
topwbatsman1

TopWomenBatsman1 = list(topwbatsman1)[:1]
TopWomenBatsman1
```

```python
topwbowler1 = []

for tpbwlr1 in top10womenbatsman1:
    tpbwlr1 = tpbwlr1.get_text()
    topwbowler1.append(tpbwlr1)
topwbowler1

TopWomenBowler1 = list(topwbowler1)[1:2]
TopWomenBowler1
```

```python
top10womenbatsman2 = soup.find_all('td', class_="table-body__cell name")
top10womenbatsman2
```

```python
topwbatsman2to10 = []

for tpbats2to10 in top10womenbatsman2:
    tpbats2to10 = tpbats2to10.get_text().split('\n')[1]
    topwbatsman2to10.append(tpbats2to10)
topwbatsman2to10

TopWomenBatsman2to10 = list(topwbatsman2to10)[:9]
TopWomenBatsman2to10
```

```python
topwbowlers2to10 = []

for tpbwlrs2to10 in top10womenbatsman2:
    tpbwlrs2to10 = tpbwlrs2to10.get_text().split('\n')[1]
    topwbatsman2to10.append(tpbwlrs2to10)
topwbowlers2to10

TopWomenBowlers2to10 = list(topwbatsman2to10)[9:18]
TopWomenBowlers2to10
```

```python
Top10WomenBatsman = (TopWomenBatsman1+TopWomenBatsman2to10)
Top10WomenBatsman
```

```python
Top10WomenBowlers = (TopWomenBowler1+TopWomenBowlers2to10)
Top10WomenBowlers
```

```python
top10womenbatsman3 = soup.find_all('div', class_="rankings-block__banner--nationality")
top10womenbatsman3
```

```python
topwbatsmancountry1 = []

for tpbatscountry1 in top10womenbatsman3:
    tpbatscountry1 = tpbatscountry1.get_text().split('\n')[2]
    topwbatsmancountry1.append(tpbatscountry1)
topwbatsmancountry1

topwomenbatsmancountry = list(topwbatsmancountry1)[:1]
topwomenbatsmancountry
```

```python
topwbowlercountry1 = []

for tpbowlscountry1 in top10womenbatsman3:
    tpbowlscountry1 = tpbowlscountry1.get_text().split('\n')[2]
    topwbowlercountry1.append(tpbowlscountry1)
topwbowlercountry1

topwomenbowlercountry = list(topwbowlercountry1)[1:2]
topwomenbowlercountry
```

```python
topwbatsmanrating1 = []

for tpbatsrating1 in top10womenbatsman3:
    tpbatsrating1 = tpbatsrating1.get_text().split('\n')[3]
    topwbatsmanrating1.append(tpbatsrating1)
topwbatsmanrating1

topwomenbatsmanrating = list(topwbatsmanrating1)[:1]
topwomenbatsmanrating
```

```python
topwbowlsmanrating1 = []

for tpbowlsrating1 in top10womenbatsman3:
    tpbowlsrating1 = tpbowlsrating1.get_text().split('\n')[3]
    topwbowlsmanrating1.append(tpbowlsrating1)
topwbowlsmanrating1

topwomenbowlersrating = list(topwbowlsmanrating1)[1:2]
topwomenbowlersrating
```

```python
top10womenbatsman4 = soup.find_all('span', class_="table-body__logo-text")
top10womenbatsman4
```

```python
topwbatsmancountry2to10 = []

for topwbats2to10 in top10womenbatsman4:
    topwbats2to10 = topwbats2to10.get_text()
    topwbatsmancountry2to10.append(topwbats2to10)
topwbatsmancountry2to10

topwomenbatsmancountry2to10 = list(topwbatsmancountry2to10)[:9]
topwomenbatsmancountry2to10
```

```python
Top10WomensBatsman = (topwomenbatsmancountry+topwomenbatsmancountry2to10)
Top10WomensBatsman
```
