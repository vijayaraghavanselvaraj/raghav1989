```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.bewakoof.com/women-tshirts?ga_q=tshirts'
```

```python
bewakoof = requests.get(url)
bewakoof
```

```python
bewakoof.content
soup = BeautifulSoup(bewakoof.content, 'html.parser')
print(soup.prettify())
```

```python
bewakoof = soup.find_all('div', class_='productCardDetail')
bewakoof
```

```python
Productname = []

for name in bewakoof:
    name = name.get_text().split('')[0]
    Productname.append(name)
Productname
```

```python
bewakoof = soup.find_all('span', class_='discountedPriceText')
bewakoof
```

```python
Productprice = []

for price in bewakoof:
    price = price.get_text()
    Productprice.append(price)
Productprice
```

```python
bewakoof = soup.find_all('img', class_='productImgTag')
bewakoof
```

```python
Images = []

for imagelinks in bewakoof:
    imagelinks=(imagelinks['src'])
    Images.append(imagelinks)
Images
```

```python
data = pd.DataFrame()
data['Product Name'] = Productname
data['Product Price'] = Productprice
data['Product Image'] = Images
data.head(11)
```

```python
data.to_csv('Bewakoof Details', index=False)
```
