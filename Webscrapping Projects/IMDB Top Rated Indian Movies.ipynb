```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.imdb.com/india/top-rated-indian-movies/'
```

```python
imdp_topim = requests.get(url)
imdp_topim
```

```python
imdp_topim.content
soup = BeautifulSoup(imdp_topim.content, 'html.parser')
print(soup.prettify())
```

```python
imdp_topim = soup.find_all('td', class_="titleColumn")
imdp_topim
```

```python
Ranking = []

for rank in imdp_topim:
    rank = rank.get_text(strip=True).replace('\n', "").split(".")[0]
    Ranking.append(rank)
Ranking
```

```python
Movies = []

for movie in imdp_topim:
    movie = movie.get_text().split('\n')[2].lstrip()
    Movies.append(movie)
Movies
```

```python
Year = []

for year in imdp_topim:
    year = year.get_text().split('\n')[3].strip("()")
    Year.append(year)
Year
```

```python
imdp_topimrating = soup.find_all('td', class_="ratingColumn imdbRating")
imdp_topimrating
```

```python
Rating = []

for rate in imdp_topimrating:
    rate = rate.get_text().replace('\n', "")
    Rating.append(rate)
Rating
```

```python
data = pd.DataFrame()

data['Ranking'] = Ranking
data['Movies'] = Movies
data['Year'] = Year
data['Rating'] = Rating
data.head(101)
```

```python
data.to_csv('Top Indian Movies in IMDB', index=False)
```
