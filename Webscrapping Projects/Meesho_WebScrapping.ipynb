```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://meesho.com/bagsladies/pl/p7vbp'
```

```python
meesho_bags = requests.get(url)
meesho_bags
```

```python
meesho_bags.content
soup = BeautifulSoup(meesho_bags.text, "html.parser")
print(soup.prettify())
```

```python
meesho_bags = soup.find_all('div', class_="Card__BaseCard-sc-b3n78k-0 dUNFgg NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS")
meesho_bags
```

```python
Name = []

for name in meesho_bags:
    name = name.get_text().split("")[0]
    Name.append(name)
Name
```

```python
meesho_bagsprice = soup.find_all('h5', class_="Text__StyledText-sc-oo0kvp-0 hiHdyy")
meesho_bagsprice
```

```python
Price = []

for price in meesho_bagsprice:
    price = price.get_text()
    Price.append(price)
Price
```

```python
meesho_discounts= soup.find_all('span', class_="Text__StyledText-sc-oo0kvp-0 lnonyH")
meesho_discounts
```

```python
Discounts = []

for discount in meesho_discounts:
    discount = discount.get_text()
    Discounts.append(discount)
Discounts
```

```python
data = pd.DataFrame()
data ['Name'] = Name
data ['Price'] = Price
data ['Discounts'] = Discounts
data.head(40)
```

```python
data.to_csv('Meesho WebScrapped Data', index=False)
```
