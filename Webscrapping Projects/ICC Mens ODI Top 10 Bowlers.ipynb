```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi'
```

```python
topodibowlers = requests.get(url)
topodibowlers
```

```python
topodibowlers.content
soup = BeautifulSoup(topodibowlers.content, 'html.parser')
print(soup.prettify())
```

```python
topodibowlers1 = soup.find_all('div', class_="rankings-block__banner--player-info")
topodibowlers1
```

```python
ODIBowlers1 = []

for odibowler1name in topodibowlers1:
    odibowler1name = odibowler1name.get_text().split('\n')[1]
    ODIBowlers1.append(odibowler1name)
ODIBowlers1

topodibowlername = list(ODIBowlers1)[1:2]
topodibowlername
```

```python
topodibowlers2 = soup.find_all('td', class_="table-body__cell name")
topodibowlers2
```

```python
ODIBowlers2to10 = []

for odibowler2to10name in topodibowlers2:
    odibowler2to10name = odibowler2to10name.get_text().split('\n')[1]
    ODIBowlers2to10.append(odibowler2to10name)
ODIBowlers2to10

topodibowlername2to10 = list(ODIBowlers2to10)[9:18]
topodibowlername2to10
```

```python
top10odibowlersname = (topodibowlername+topodibowlername2to10)
top10odibowlersname
```

```python
topodibowlers3 = soup.find_all('div', class_="rankings-block__banner--player-info")
topodibowlers3
```

```python
ODIBowlerscountry1 = []

for odibowler1country in topodibowlers3:
    odibowler1country = odibowler1country.get_text().split('\n')[4]
    ODIBowlerscountry1.append(odibowler1country)
ODIBowlerscountry1

TopODIBowlercountry1 = list(ODIBowlerscountry1)[1:2]
TopODIBowlercountry1
```

```python
ODIBowlersRating1 = []

for odibowler1rating in topodibowlers3:
    odibowler1rating = odibowler1rating.get_text().split('\n')[5]
    ODIBowlersRating1.append(odibowler1rating)
ODIBowlersRating1

TopODIBowlerRating1 = list(ODIBowlersRating1)[1:2]
TopODIBowlerRating1
```

```python
topodibowlers4 = soup.find_all('span', class_="table-body__logo-text")
topodibowlers4
```

```python
ODIBowlerscoutry2to10 = []

for odibowler2to10country in topodibowlers4:
    odibowler2to10country = odibowler2to10country.get_text()
    ODIBowlerscoutry2to10.append(odibowler2to10country)
ODIBowlerscoutry2to10

TopBowlersCountry2to10 = list(ODIBowlerscoutry2to10)[9:18]
TopBowlersCountry2to10
```

```python
top10odibowlerscountry = (TopODIBowlercountry1+TopBowlersCountry2to10)
top10odibowlerscountry
```

```python
topodibowlers5 = soup.find_all('td', class_="table-body__cell u-text-right rating")
topodibowlers5
```

```python
ODIBowlersRating2to10 = []

for odibowler2to10rating in topodibowlers5:
    odibowler2to10rating = odibowler2to10rating.get_text()
    ODIBowlersRating2to10.append(odibowler2to10rating)
ODIBowlersRating2to10

TopBowlersRating2to10 = list(ODIBowlersRating2to10)[9:18]
TopBowlersRating2to10
```

```python
TopBowlersRating = (TopODIBowlerRating1+TopBowlersRating2to10)
TopBowlersRating
```

```python
data = pd.DataFrame()
data['Name'] = top10odibowlersname
data['Country'] = top10odibowlerscountry
data['Rating'] = TopBowlersRating
data.head(20)
```

```python
data.to_csv ('ICC Mens ODI Top 10 Bowlers', index=False)
```
