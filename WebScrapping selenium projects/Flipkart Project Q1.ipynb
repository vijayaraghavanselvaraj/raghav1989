```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
url = "https://www.flipkart.com/"
driver.get(url)
```

```python
prod_search = driver.find_element_by_class_name('_3704LK')
prod_search
```

```python
prod_search.send_keys('sunglasses')
prod_search
```

```python
submit_button = driver.find_element_by_xpath('/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
submit_button
```

```python
submit_button.click()
```

```python
prod_name1 = driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
prod_name1
```

```python
Brandname = []

for name in prod_name1:
    Brandname.append(name.text)
Brandname[:100]
```

```python
prod_desc1 = driver.find_elements_by_xpath('//a[@class="IRpwTa"]')[:31]
prod_desc2 = driver.find_elements_by_xpath('//a[@class="IRpwTa _2-ICcC"]')
prod_desc3 = driver.find_elements_by_xpath('//a[@class="IRpwTa"]')[31:]
prod_desc1

product_description = list(prod_desc1+prod_desc2+prod_desc3)
product_description
```

```python
Description = []

for desc in product_description:
    Description.append(desc.text)
Description[:100]
```

```python
prod_price1 = driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
prod_price1
```

```python
Price = []

for rate in prod_price1:
    Price.append(rate.text)
Price[:100]
```

```python
data = pd.DataFrame()
data ['Brandname'] = Brandname
data ['Price'] = Price
data ['Description'] = Description
data.head(100)
```
