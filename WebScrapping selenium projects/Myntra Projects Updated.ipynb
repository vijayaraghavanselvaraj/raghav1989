```python
import pandas as pd
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
```

```python
myntra = webdriver.Chrome('chromedriver.exe')
```

```python
url = "https://www.myntra.com/"
myntra.get(url)
```

```python
search = myntra.find_element_by_xpath('//input[@class="desktop-searchBar"]')
search
```

```python
search.send_keys('shoes')
```

```python
pricefilter = myntra.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
pricefilter
```

```python
shoeslistname = myntra.find_elements_by_xpath('//h3[@class="product-brand"]')
shoeslistname
```

```python
shoesbrandname = []

for name in shoeslistname:
    shoesbrandname.append(name.text)
shoesbrandname
```

```python
shoesdesc = myntra.find_elements_by_xpath('//h4[@class="product-product"]')
shoesdesc
```

```python
shoesdescription = []

for desc in shoesdesc:
    shoesdescription.append(desc.text)
shoesdescription
```

```python
shoespricelist = myntra.find_elements_by_xpath('//div[@class="product-price"]')
shoespricelist
```

```python
shoeprice = []

for price in shoespricelist:
    shoeprice.append(price.text.split('Rs.')[1])
shoeprice
```

```python
data = pd.DataFrame()
data ['Brand Name'] = shoesbrandname
data ['Price'] = shoeprice
data ['Description'] = shoesdescription
data.head(101)
```
